```python
from collections import Counter

################################
#STANDARDIZING THE UNIQUE WORDS#
################################
    
def append_regularized_unique_words(old_df):
    old_df['std_unique_words'] = old_df.groupby(['essay_set'])[['unique_words']].apply(lambda x: (x - np.mean(x)) / (np.std(x)))
    return old_df

def create_regularization_unique_words(old_df):
    #getting the number of datasets
    max_essay_set = max(old_df['essay_set'])
    #list of the regularized values
    regularization_data = []
    for i in range(max_essay_set+1):
        mean = np.mean((old_df[old_df['essay_set'] == i + 1])['unique_words'])
        std = np.std((old_df[old_df['essay_set'] == i + 1])['unique_words'])
        regularization_data.append([i + 1, mean, std])
    return regularization_data

################################
#STANDARDIZING THE TOTAL WORDS#
################################

def append_regularized_total_words(old_df):
    old_df['std_total_words'] = old_df.groupby(['essay_set'])[['total_words']].apply(lambda x: (x - np.mean(x)) / (np.std(x)))
    return old_df

def create_regularization_total_words(old_df):
    #getting the number of datasets
    max_essay_set = max(old_df['essay_set'])
    #list of the regularized values
    regularization_data = []
    for i in range(max_essay_set+1):
        mean = np.mean((old_df[old_df['essay_set'] == i + 1])['total_words'])
        std = np.std((old_df[old_df['essay_set'] == i + 1])['total_words'])
        regularization_data.append([i + 1, mean, std])
    return regularization_data

#####################################################
#COUNTING THE NUMBER OF UNIQUE WORDS AND TOTAL WORDS#
#####################################################

def fill_unique_words_column(train_df, valid_df):

    #percentage of unique words to the total number of words
    unique_word_percentages_train = []
    unique_word_percentages_valid = []

    for i in range(len(train_df)):
        splits = train_df.iloc[i]["essay"].split()
        total_words = len(splits)
        unique_words = len(Counter(splits))
        percentage = float(unique_words) / total_words
        unique_word_percentages_train.append(percentage)
        
    for i in range(len(valid_df)):
        splits = valid_df.iloc[i]["essay"].split()
        total_words = len(splits)
        unique_words = len(Counter(splits))
        percentage = float(unique_words) / total_words
        unique_word_percentages_valid.append(percentage)    

    #Add the features to the dataset
    train_df["unique_words"] = unique_word_percentages_train
    valid_df["unique_words"] = unique_word_percentages_valid

    #Appending standardized unique words to training set
    append_regularized_unique_words(train_df)
    
def fill_total_words_column(train_df, valid_df):
    #Total number of words
    total_number = []
    
    for i in range(len(train_df)):
        splits = train_df.iloc[i]["essay"].split()
        total_words = len(splits)
        total_number.append(total_words)
        
    train_df["total_words"] = total_number 
    
    #Appending standardized total words to training set
    append_regularized_total_words(train_df)
```
